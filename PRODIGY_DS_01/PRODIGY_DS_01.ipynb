Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

````

````python
data_file_path = 'population data.csv'
population_data = pd.read_csv(data_file_path)
population_data.head()
````

````python
# Select the year 2020 and relevant columns
year = '2020'
population_2023 = population_data[['Country Name', year]]
# Sort the data for better visualization
population_2023_sorted = population_2023.sort_values(by=year, ascending=False).head(10)  # Top 10 countries by population

# Create a bar chart
plt.figure(figsize=(14, 8))
sns.set_palette("pastel")
plt.bar(population_2023_sorted['Country Name'], population_2023_sorted[year])
plt.xlabel('Country' )
plt.ylabel('Population in 2023')
plt.title('Top 10 Countries by Population in 2023', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
````
